# Solution

```python
# Read input
P = int(input())  # Number of packages delivered
C = int(input())  # Number of collisions with obstacles

# Calculate score
score = P * 50 - C * 10  # Points gained from packages delivered and lost from collisions

# Apply bonus if applicable
if P > C:
    score += 500

# Print the final score
print(score)
```

```python
# Read input
R = int(input())  # Number of regular boxes
S = int(input())  # Number of small boxes

# Calculate total number of cupcakes
total_cupcakes = R * 8 + S * 3

# Calculate number of cupcakes left over
left_over_cupcakes = total_cupcakes - 28

# Print the number of cupcakes left over
print(left_over_cupcakes)
```
