```python
def calculate_spiciness(peppers_list):
    # Define a dictionary with the SHU values of the peppers
    shu_values = {
        "Poblano": 1500,
        "Mirasol": 6000,
        "Serrano": 15500,
        "Cayenne": 40000,
        "Thai": 75000,
        "Habanero": 125000
    }

    total_spiciness = 0

    # Calculate the total spiciness by adding up the SHU values of the peppers
    for pepper in peppers_list:
        total_spiciness += shu_values[pepper]

    return total_spiciness

# Read input
n = int(input())
peppers_list = [input() for _ in range(n)]

# Calculate and print the total spiciness
print(calculate_spiciness(peppers_list))
```

```python
def calculate_star_rating(points, fouls):
    return points * 5 - fouls * 3

# Read input
n = int(input())
players_greater_than_40 = 0
gold_team = True

# Calculate star ratings and count players with rating greater than 40
for _ in range(n):
    points = int(input())
    fouls = int(input())

    star_rating = calculate_star_rating(points, fouls)
    
    if star_rating > 40:
        players_greater_than_40 += 1
    else:
        gold_team = False

# Output the results
print(players_greater_than_40, end='')
if gold_team:
    print('+')
else:
    print()
```

```python
# Read the input
n = int(input())

while n != 1:
    if n % 2 == 0:
        n = n // 2
        print(n, end=" ")
    else:
        n = n * 3 + 1
        print(n, end=" ")
```
